```python
import numpy as np

def metodo_simpson(I, a, b, n):
    if n % 2 != 0:
        print("El número de subdivisiones n debe ser par.")
    else:
        h = (b - a) / n  
        x = np.linspace(a, b, n + 1) 
        y = I(x)  # Valores de la función en los puntos x
        # Aplicar la regla de Simpson: suma ponderada de y
        suma = y[0] + y[-1]  # Primer y último término (peso 1)
    
        for i in range(1, n):
            if i % 2 == 0:
                suma += 2 * y[i]  # Términos pares (peso 2)
            else:
                suma += 4 * y[i]  # Términos impares (peso 4)
        integral = (h / 3) * suma
    return integral

# Ejemplo de uso:
def I(x):
    return x**2 + 3 * np.sin(x)  # Función a integrar

a = 0  # Límite inferior
b = 2  # Límite superior
n = 100  # Número de subdivisiones (debe ser par)

resultado = metodo_simpson(I, a, b, n)
print(f"Aproximación de la integral usando Simpson: {resultado:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Simpson
a_s = 0
b_s = 1
n_s = 1000
h_s = (b_s-a_s)/n_s
x_s = np.linspace(a_s, b_s, n_s+1)
T = np.linspace(0.1, 100, 10000)

V_s=[]

for t in T:
    def I(x_s):
        return (t**4/(2*np.pi**2))*x_s**2*np.log(1-np.exp(-(x_s**2+(0.04/t)**2)**(1/2)))

    y_s = I(x_s)
    suma=0
    
    if n_s % 2 != 0:
        print('Las subdiviciones no son pares')
    else:
        extremos = y_s[0]+ y_s[-1]
        
        #todo lo de enmedio
        for i in range(1,n_s):
            if i % 2 == 0:
                suma = suma + 2 * y_s[i] 
            else:
                suma = suma + 4 * y_s[i] 
                
    integral= (h_s/3)*(extremos + suma)
    V_s.append(integral)

V_s = np.array(V_s)

plt.plot(T, V_s ,label='Resultado Simpson', color='cyan')
plt.title('Correción un lazo')
plt.xlabel('T')
plt.ylabel('V')
plt.grid(True)
plt.legend()
plt.show()
```
