# Sección nueva

```python
#Celda donde se hace un gráfico 3D de lamda, m y g.
# --------------------------------------------------------------------------

df['lamda'] = df['lamda'].astype(str)    # Convertir 'lamda' a un data frame

# Creamos el gráfico de dispersión.
fig = px.scatter_3d(df, x='lamda', y='m', z='g', color='lamda',   #Ajuste de los ejes.
                    title='Gráfico 3D de λ, m y g',
                    labels={'lamda': 'λ', 'm': 'm', 'g': 'g'})    #Nombre de los ejes

fig.update_traces(marker=dict(size=5, opacity=0.5)) # Tamaño de los puntos y la opacidad
fig.update_layout(width=1000, height=800)           #Anchura y Tamaño
fig.show()
```

```python

#Gráficos de 2D de m vs g para distintos valores de lamda
# --------------------------------------------------------

lamda_values = df['lamda'].unique()     # Valores de lamda

# --------------------------------------------------------
#Diseño de los subgráficos (Como lo muestra)

n_lamda = len(lamda_values)     # Número de lamdas
n_cols = 3  # Número de columnas
n_rows = (n_lamda + n_cols - 1) // n_cols  # Calcular filas

# Crear una figura con subgráficos
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows), squeeze=False)
axes = axes.flatten()  # Aplanar los axes para iterar fácilmente

# Graficar para cada valor de lamda
for idx, lamda_val in enumerate(lamda_values):
    df_lamda = df[df['lamda'] == lamda_val]     # Filtrar por el valor de lamda

    axes[idx].scatter(df_lamda['m'], df_lamda['g'], alpha=0.6, s=50)
    axes[idx].set_title(f'λ = {lamda_val}')
    axes[idx].set_xlabel('m')
    axes[idx].set_ylabel('g')
    axes[idx].grid(True, linestyle='--', alpha=0.7)
    axes[idx].set_ylim(0.1,5)
    axes[idx].set_xlim(50, 150)

for idx in range(n_lamda, len(axes)):       # Eliminar subgráficos vacíos
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()
```

```python
# Graficos 2D de "lamda vs m"    y   "lamda vs g"
# -----------------------------------------------

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))   # Figura con 1 fila y 2 columnas

# Gráfico 1: lamda vs m
ax1.scatter(df['lamda'], df['m'], alpha=0.6, s=50, c='blue')
ax1.set_xlabel('λ')
ax1.set_ylabel('m')
ax1.set_title('λ vs m')
ax1.grid(True, linestyle='--', alpha=0.7)

# Gráfico 2: lamda vs g
ax2.scatter(df['lamda'], df['g'], alpha=0.6, s=50, c='green')
ax2.set_xlabel('λ')
ax2.set_ylabel('g')
ax2.set_title('λ vs g')
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
