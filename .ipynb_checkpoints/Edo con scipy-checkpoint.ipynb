```python
import numpy as np
import scipy.integrate as si
import matplotlib.pyplot as plt

```

```python
# Definir la ecuacion diferencial: dy/dx = -2y
Edo='dy/dx = -2y'
def ecuacion_dif(x,y):
    return -2 * y
```

```python
#Segunda ecuacion diferencial: dy/dx= 2y+x
Edo='dy/dx= 2y+x'
def ecuacion_dif(x,y):
    return 2*y+x
```

```python
#Tercera ecuacion diferencial: dy/dx= x/y
Edo='dy/dx= y/x'
def ecuacion_dif(x,y):
    return 2*y/x
```

```python
# Condición inicial: y(1) = 1
intervalo_x = (1, 5)  # Intervalo de tiempo
y0 = [1]
```

```python
# Resolver la EDO
sol = si.solve_ivp(ecuacion_dif, intervalo_x, y0, t_eval=np.linspace(1, 5, 100))
```

```python
# Graficar la solución
plt.plot(sol.t, sol.y[0])
plt.title('Solución de la Ecuacion diferencial '+Edo)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```
